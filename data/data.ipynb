```python
#Imports
import pandas as pd
from pathlib import Path

#Load and display data
data_to_load = Path('C:/Users/victo/Documents/Bootcamp/Week16/M16-project-3-RTT/data/Datafiniti_Fast_Food_Restaurants_May19.csv')
restaurant_data = pd.read_csv(data_to_load)

restaurant_data.head()
```

```python
#Drop unwanted columns
restaurant_data = restaurant_data.drop(columns=['id', 'dateAdded', 'dateUpdated', 'primaryCategories', 'keys', 'sourceURLs'])

restaurant_data.head()
```

```python
#Create list from categories column
categories_word_list = restaurant_data['categories'].values.tolist()
```

```python
categories_word_list
```

```python
clean_categories_word_list = []

for category in categories_word_list:
    clean_category = category.replace(',', '').replace(' and ', '').replace('Restaurants', '').replace('Restaurant', '').replace('Take Out', '').replace('Fast Food', '').replace('Carry-out', '').replace('food', ''),
    clean_categories_word_list.append(clean_category)

clean_categories_word_list

```
